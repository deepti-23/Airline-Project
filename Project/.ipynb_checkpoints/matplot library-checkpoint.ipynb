```python
# sales data for car company
sales=[2014,2015,2016,2017,2018,2019]
sales_count=[39,120,115,112,75,30]
sales_enquiry=[0,0,15,60,40,20]

#mymarketdata={'CarCompany':['Toyata','Toyata','Hyundai','Hyundai','Tata','Tata'],
#              'Customers' :['Deepti','Udit','Prakhar','suman','Ananya','Varun'],
#              'Expenses'  :[1000,2000,3000,4000,5000,6000]}
```

```python
pl.plot(sales,sales_count)
pl.xlabel('year wise car sales')
pl.ylabel('no of cars sold year wise')
```

```python
pl.plot(sales,sales_enquiry,color='yellow')
pl.plot(sales,sales_count)
pl.xlabel('year wise car sales')
pl.ylabel('Number')
```

```python
pl.bar(sales,sales_count)
pl.xlabel('year wise car sales')
pl.ylabel('no of cars sold year wise')
```

```python
pl.hist(sales,bins=10)
```

```python
pl.scatter(sales,sales_count)
```

```python
pl.scatter(sales,sales_enquiry,color='green')
pl.scatter(sales,sales_count,color='purple')
pl.xlabel('year wise car sales')
pl.ylabel('no of cars sold year wise')
```

```python
# Area plot
#  employee working schedule
days=[1,2,3,4,5]
worktime=[7,8,10,6,9]
extrahours=[3,4,2,3,1]
shorttime=[1,2,0,1,3]
breaktime=[1,2,4,0,5]

pl.stackplot(days,worktime,extrahours,shorttime,breaktime,colors=['b','g','y','r'])
```

```python

# PIE CHART 

carbrands = 'TATA','TOYOTA','HYUNDAI','VALKSWAGEN','FORD'
customers = [20,10,30,15,25]
companycolors=['gold','silver','blue','red','yellow']
listexp = (0,0,.3,0,0)
#pl.pie(customers,labels=carbrands)
#pl.pie(customers,labels=carbrands,colors=companycolors,autopct='%1.0f')
pl.pie(customers,labels=carbrands,colors=companycolors,autopct='%1.0f',explode=listexp)
```
